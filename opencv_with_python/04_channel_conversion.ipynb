## Image Channel Convresion

```python
# import all the lib
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn.metrics as sm
import matplotlib.pyplot as plt
import cv2
import sys
%matplotlib inline
```

```python
# Import the image color change due to matplot and cv2 color channel
img = cv2.imread('data/img.jpg')
plt.imshow(img)
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray')
plt.show()
```

# Converting to HSV and HSL

```python
# Transform the image into HSV and HLS models
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

## Convert Image into WB or Binary Image

```python
import cv2
# 1. Read a grayscale image
im_gray = cv2.imread('data/img.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# 2. Convert grayscale image to binary
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

which determines the threshold automatically from the image using Otsu's method, or if you already know the threshold you can use:

```python
thresh = 127
im_bw = cv2.threshold(im_gray, thresh, 255, cv2.THRESH_BINARY)[1]
```

```python
# Save to disk

cv2.imwrite('bw_image.png', im_bw)
plt.imshow(im_bw)
```
