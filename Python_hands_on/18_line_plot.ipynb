```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pholl = sns.load_dataset("titanic")
pholl.head()
```

```python
sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
plt.title("Flower Plot")
plt.show()
```

```python
sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
plt.title("Flower Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
sns.set_style(style= "darkgrid")
sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
plt.title("Flower Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
# fig size
pholl = sns.load_dataset("iris")
plt.figure(figsize=(8,6))
sns.set_style(style= "darkgrid")
sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
plt.title("Flower Plot")
plt.show()
```
