## Image Channel Convresion to Black and White

```python
# import all the lib
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn.metrics as sm
import matplotlib.pyplot as plt
import cv2
import sys
%matplotlib inline
```

```python
# Import the image color change due to matplot and cv2 color channel
img = cv2.imread('data/img.jpg')
plt.imshow(img)
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

## Saving a Gray Image

```python
import cv2
im_gray = cv2.imread('data/img.jpg', cv2.IMREAD_GRAYSCALE)
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
thresh = 127
im_bw = cv2.threshold(im_gray, thresh, 255, cv2.THRESH_BINARY)[1]
# Save to disk

cv2.imwrite('data/bw_image.png', im_bw)
plt.imshow(im_bw)
```
