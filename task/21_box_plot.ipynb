```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pholl = sns.load_dataset("iris")

sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
sns.set_style('dark')
plt.title("Flower Line Plot")
plt.show()
```

```python
kashti = sns.load_dataset("titanic")
print(kashti.head(5))
sns.barplot(x='fare', y = "class", data=kashti, saturation=0.2)
sns.set_style('dark')
plt.title("Flower Line Plot")
plt.show()
```

```python
# load the two necessary column
df = sns.load_dataset('titanic').loc[:, ['sex', 'survived']]

# create a pivot table
dfp = df.pivot_table(index='sex', columns=['survived'], aggfunc=len)

# plot the dataframe
dfp.plot(kind='bar', stacked=True, ylabel='Counts', xlabel='Gender',
         title='Survival Status Count by Gender', rot=0)
```

```python
# errcolor='0.2' 0 balck and 1 is white
# edgecolor = '0.2' 0 to one
sns.barplot(x='class', y = "fare", data=kashti, linewidth = 2.5, facecolor = (0,0,0,1), errcolor='1', edgecolor = '1')
```
