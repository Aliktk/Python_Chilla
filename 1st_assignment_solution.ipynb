# Importing libraries

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pholl = sns.load_dataset("iris")
pholl.head()
```

```python
sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
plt.title("Flower Line Plot")
plt.show()
```

# How to change the background color of the graph

```python
# Use the seaborn.set() Function to Change the Background Color of Seaborn Plots in Python
# Use the seaborn.set_style() Function to Change the Background Color of Seaborn Plots in Python
#  white, dark, whitegrid, darkgrid, ticks
sns.set_style("darkgrid")
sns.lineplot(x='sepal_length', y = "sepal_width", data=pholl)
plt.title("Flower Line Plot")
plt.show()
```

```python
kashti = sns.load_dataset("titanic")
kashti.to_csv("titanic.csv")
kashti.head(2)
```

```python
kashti = sns.load_dataset("titanic")
sns.set_style("darkgrid")
sns.lineplot(x='who', y = "alive", hue= 'sex', data=kashti)
plt.title("Flower Line Plot")
plt.show()
```
